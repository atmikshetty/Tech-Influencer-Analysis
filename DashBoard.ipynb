```python
# Importing the necessary libraries
import pandas as pd
from ntscraper import Nitter
```

```python
# Initializing the scraper instance(breaks down sometimes due to the rate limit or the instance being restricted)
scraper = Nitter()
```

```python
# Function to scrape only the required data from the influencer's accounts
def get_tweets(name, modes, no):
  tweets = scraper.get_tweets(name, mode=modes, number=no)  #'get_tweets' provided by ntscraper
  final_tweets = []
  for tweet in tweets['tweets']:
    # Filtering out only the necessary columns
    data = [tweet['link'], tweet['text'], tweet['user']['username'], tweet['date'], tweet['stats']['likes'], tweet['stats']['comments'], tweet['stats']['retweets']]
    final_tweets.append(data)
  data = pd.DataFrame(final_tweets, columns = ['link', 'text', 'username', 'date', 'likes', 'comments', 'retweets'])  # Creating a data frame
  return data
```

```python
# Scraping data for Aman Dhattarwal, other two influencers data has already been scraped and added with the source code on github
data = get_tweets('AmanDhattarwal', 'user', 100)
```

## 3. Filtering out the tweets with promotions and marketing for the respective courses

Since the purpose of this project is to show the reach of promotional vs non-promotional tweets, due to less data, the results aren't exceptional but still some insights can be generated.

Steps Performed:

* Creating an array of the words frequently used in promotional posts like 'discount', 'course', etc.
* Lambda function to create a new column which depicts whether the tweet is a promotional tweet or not
* Summary statistics of User Engagement

```python
# Words frequently used in promotional posts
words = ['a2z', 'sde sheet', 'blind 75', 'alpha', 'sigma', 'cohort', 'course', 'paid', 'placement', 'dev', 'dsa', 'discount']
```

```python
# Filtering out the promotional tweets
data['promotion'] = data['text'].apply(lambda x: any(word in x.lower() for word in words))
```

```python
# Summary Statistics of the user engagement
summary = data.groupby('promotion').agg({
    'likes': 'mean',
    'retweets': 'mean',
    'comments': 'mean'
}).reset_index()
```

## 4. Matplotlib and Seaborn Visaulization of Promotional vs Non-Promotional Tweets



```python
# Importing the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Countplot for Number of Promotional vs Non-Promotional Tweets
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
palette = sns.color_palette("Set2")

ax = sns.countplot(x='promotion', data=data, palette=palette)

plt.title('Number of Promotional vs Non-Promotional Tweets', fontsize=16)
plt.xlabel('Tweet Type', fontsize=14)
plt.ylabel('Count', fontsize=14)
ax.set_xticklabels(['Non-Promotional', 'Promotional'], fontsize=12)

# Annotation of data labels
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 9), textcoords='offset points', fontsize=12)
plt.show()

"""
As mentioned before, the data split is not suitable for a proper visualization but due to rate limits, the dashboard had to be built with this itself. Apologies.
"""
```

```python
# Swarm plot for individual data points
plt.figure(figsize=(12, 6))

sns.set_style("whitegrid")
palette = sns.color_palette("muted")

sns.violinplot(x='promotion', y='likes', data=data, palette=palette, inner='quartile', split=True)
sns.swarmplot(x='promotion', y='likes', data=data, color='k', alpha=0.5)

plt.title('Likes Distribution for Promotional vs Non-Promotional Tweets', fontsize=16)
plt.xlabel('Tweet Type', fontsize=14)
plt.ylabel('Likes', fontsize=14)

plt.xticks(ticks=[0, 1], labels=['Non-Promotional', 'Promotional'], fontsize=12)
plt.show()


"""
The likes for Non-promotional tweets are quite high, this could be due to lot of factors, on manual checking majority of the most liked tweets of these influencers
are from meet-ups, motivational content and their day to day office work.
"""
```

```python
"""Harkirat Singh's Twitter is filled with majority of posts related to his latest course. So we see how many of the tweets consist the promotion for his course"""

# Checking for word cohort
data['cohort'] = data['text'].str.contains('cohort', case=False, na=False)

cohort_counts = data[data['cohort'] == True].groupby('username').size().reset_index(name='count')
cohort_counts = cohort_counts.sort_values(by='count', ascending=False)

# Plotting a bar plot
plt.figure(figsize=(15, 8))

sns.set_style("whitegrid")
ax = sns.barplot(x='count', y='username', data=cohort_counts, palette=palette)

plt.title('Number of Tweets Containing the Word "Cohort" by Username', fontsize=16)
plt.xlabel('Number of Tweets', fontsize=14)
plt.ylabel('Username', fontsize=14)

for p in ax.patches:
    ax.annotate(f'{int(p.get_width())}', (p.get_width(), p.get_y() + p.get_height() / 2),
                ha='center', va='center', xytext=(5, 0), textcoords='offset points', fontsize=12)

plt.show()

""" Usernames and all the tweets reposted by Harkirat regarding his cohort.
Surprisingly none of the tweets seem to have been posted by Harkirat since all the 100 latest tweets scraped are retweets.
This shows a positive response regarding his course but a point to notice is that none of these users in the short time span have shared updates regarding the
course except one. This could say that either they have quit the course or have stopped sharing updates on it.
The second should not be the case since most of his students are told from the start to post on twitter and follow the #buildinpublic trend.
"""
```

## Times Series Forecasting on the data to figure out during which months the no of tweets sky rocketed.

An assumption here after going through the influencers social media would be from **January** since influencers usually launch their courses around that time and also Aman and Harkirat launched their courses around that time and Striver also started updating his free DSA course.

```python
# Converting the data to the required format
data['date'] = pd.to_datetime(data['date'], format='%b %d, %Y · %I:%M %p %Z')
```

```python
data.set_index('date', inplace=True)
```

```python
data['tweet_count'] = 1
monthly_tweets = data['tweet_count'].resample('M').sum()
```

```python
# Line plot for the forecasting
plt.figure(figsize=(14, 8))

sns.set_style("whitegrid")

ax = monthly_tweets.plot(color='b', marker='o', linestyle='-', linewidth=2, markersize=6)
plt.title('Tweet Frequency Over Time', fontsize=20)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Number of Tweets', fontsize=16)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
ax.legend(['Monthly Tweet Count'], fontsize=14)

plt.show()

"""
We can see how during January the posts have exploded massively as compared to other months.
"""
```

```python
# A simple heatmap to see how the engagement metrics affect one another
plt.figure(figsize=(10, 8))
sns.heatmap(data[['likes', 'retweets', 'comments']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Engagement Metrics')
plt.show()
```

## 5. Plotly(Dash) Dashboard

**Dash** is used to create simple data visualisation boards, it is inbuilt and integrated with Plotly.

In the following lines of code a simple dashboard showing various insights has been created.

```python
!pip install dash
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
```

```python
# Initializing the dash app
app = dash.Dash(__name__)
```

```python
# First Visualization
fig_bar = px.histogram(data, x='promotion', color='promotion', title='Number of Promotional vs Non-Promotional Tweets',
                       labels={'promotion': 'Tweet Type', 'count': 'Count'})
```

```python
# Second Visualization
fig_violin = px.violin(data, x='promotion', y='likes', color='promotion', box=True, points='all',
                        title='Likes Distribution for Promotional vs Non-Promotional Tweets',
                        labels={'promotion': 'Tweet Type', 'likes': 'Likes'})
```

```python
# Third Visualization
data['cohort'] = data['text'].str.contains('cohort', case=False, na=False)
cohort_counts = data[data['cohort'] == True].groupby('username').size().reset_index(name='count')
cohort_counts = cohort_counts.sort_values(by='count', ascending=False)
fig_bar_cohort = px.bar(cohort_counts, x='count', y='username', orientation='h',
                        title='Number of Tweets Containing the Word "Cohort" by Username',
                        labels={'count': 'Number of Tweets', 'username': 'Username'})
```

```python
# Tweet Frequency
fig_line_forecast = px.line(monthly_tweets, title='Tweet Frequency Over Time',
                            labels={'index': 'Time', 'tweet_count': 'Number of Tweets'})
```

```python
app.layout = html.Div(children=[
    html.H1(children='Coding Influencers Tweets Analysis Dashboard'),

    # Plot 1: Countplot for Number of Promotional vs Non-Promotional Tweets
    html.Div(children=[
        html.H2(children='Countplot for Number of Promotional vs Non-Promotional Tweets'),
        dcc.Graph(id='bar-chart', figure=fig_bar)
    ]),

    # Plot 2: Violin and Swarm Plot for Likes Distribution
    html.Div(children=[
        html.H2(children='Violin and Swarm Plot for Likes Distribution'),
        dcc.Graph(id='violin-plot', figure=fig_violin)
    ]),

    # Plot 3: Bar Plot for Tweets Containing the Word "Cohort" by Username
    html.Div(children=[
        html.H2(children='Bar Plot for Tweets Containing the Word "Cohort" by Username'),
        dcc.Graph(id='cohort-bar-plot', figure=fig_bar_cohort)
    ]),

    # Plot 4: Line Plot for Tweet Frequency Over Time
    html.Div(children=[
        html.H2(children='Line Plot for Tweet Frequency Over Time'),
        dcc.Graph(id='line-plot-forecast', figure=fig_line_forecast)
    ])
])
```

```python
if __name__ == '__main__':
    app.run_server(debug=True)
```
